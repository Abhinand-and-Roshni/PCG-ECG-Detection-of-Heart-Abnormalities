```python
from sklearn import preprocessing
import pandas as pd
import numpy as np
```

```python
unscaled_df = pd.read_csv("FEATURES_FULLWAVEFORM_RRI_PHYSIO_NO NORMLAIZATION.csv")
unscaled_df.head()
```

```python
unscaled_df = unscaled_df.drop(['file_name'], axis = 1)
```

```python
unscaled_df
```

```python
df_min_max_scaled = unscaled_df
  
for column in df_min_max_scaled.columns:
    df_min_max_scaled[column] = ((df_min_max_scaled[column] - df_min_max_scaled[column].min())) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())    
  
df_min_max_scaled.head()
```

```python
df = df_min_max_scaled
```

```python
df.describe()
```

```python
df['label'] = unscaled_df['label']
df
```

```python
df.to_csv("NORMALIZED_PHYSIONET_ECG_FEATURES.csv")
```

```python
df['label'].value_counts()
```
