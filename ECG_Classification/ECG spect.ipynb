```python
import numpy as np
from scipy import signal
import matplotlib as mpl
import matplotlib.pyplot as plt
import requests  # pip install requests
%matplotlib inline
mpl.rcParams['figure.figsize'] = [12, 8]
```

```python
import scipy.io
mat = scipy.io.loadmat(f"C:/Users/Roshni/OneDrive/Desktop/Sem7/Project 1/ECG UI/TestData/A03003.mat")
```

```python
data = np.array(mat['val'])
data = np.transpose(data)
data
```

```python
list = []
for x in data:
    for i in x:
        list.append(i)
```

```python
data = list
```

```python
data = np.array(data)
```

```python
data.shape
```

```python
fig, ax = plt.subplots()
ax.plot(data)
fig = ax.get_figure()
ax.title.set_text("ECG Data")
ax.set_xlabel("Time")
ax.set_xlabel("Voltage")
fig.savefig("output.png")
```

```python
fig, ax = plt.subplots() 
f, t, Sxx = signal.spectrogram(data)
pc = ax.pcolormesh(t, f, Sxx, norm=mpl.colors.LogNorm(vmin=Sxx.min(), vmax=Sxx.max()), cmap='inferno')
ax.set_ylabel('Frequency')
ax.set_xlabel('Time') 
fig.colorbar(pc) 
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1)
ax1.plot(data)
ax1.set_title("The ECG data")
pc = ax2.pcolormesh(t, f, Sxx, cmap='inferno')
ax2.set_ylabel('Frequency')
ax2.set_xlabel('Time') 
ax2.set_title("Spectrogram for the ECG data, normal scale")
pc = ax3.pcolormesh(t, f, Sxx, norm=mpl.colors.LogNorm(vmin=Sxx.min(), vmax=Sxx.max()), cmap='inferno')
ax3.set_ylabel('Frequency')
ax3.set_xlabel('Time') 
ax3.set_title("Spectrogram for the ECG data, Log-scale")
plt.tight_layout()  # Fix spacing when using axis-level titles
```
